```python
import pandas as pd
```

```python
#import 2018 csv
data = pd.read_csv("../Resources/2018.csv")
data.head()
```

```python
data.columns
```

```python
# make df with only the important columns
cleaned = data[['YearsCodingProf','DevType','ConvertedSalary','CompanySize']]
cleaned['Year'] = '2018'
cleaned = cleaned.dropna()
```

```python
cleaned['DevType'].value_counts()
```

```python
# create smaller df with only the jobs we want
analyst = cleaned[cleaned['DevType'].str.contains('analyst')]
scientist = cleaned[cleaned['DevType'].str.contains('scientist')]
machine = cleaned[cleaned['DevType'].str.contains('machine')]
# engineer = cleaned[cleaned['DevType'].str.contains('engineer')]
```

```python
# convert job titles to specific
analyst['DevType'] = 'Data Analyst'
scientist['DevType'] = 'Data Scientist'
machine['DevType'] = 'Machine Learning Engineer'
```

```python
combined = pd.concat([analyst, scientist, machine], ignore_index=True)
combined
```

```python
combined['CompanySize']=combined['CompanySize'].replace(["10,000 or more employees"],["10000+"])
combined['CompanySize']=combined['CompanySize'].replace(["20 to 99 employees"],["10-100"])
combined['CompanySize']=combined['CompanySize'].replace(["100 to 499 employees"],["100-1000"])
combined['CompanySize']=combined['CompanySize'].replace(["1,000 to 4,999 employees"],["1000-4999"])
combined['CompanySize']=combined['CompanySize'].replace(["2 to 9 employees"],["2-9"])
combined['CompanySize']=combined['CompanySize'].replace(["10 to 19 employees"],["10-100"])
combined['CompanySize']=combined['CompanySize'].replace(["500 to 999 employees"],["100-1000"])
combined['CompanySize']=combined['CompanySize'].replace(["5,000 to 9,999 employees"],["5000-9999"])
combined['CompanySize']=combined['CompanySize'].replace(["Just me - I am a freelancer, sole proprietor, etc."],["1"])
combined['CompanySize']=combined['CompanySize'].replace(["I don’t know"],["0"])

combined
```

```python
combined['YearsCodingProf'].value_counts()
```

```python
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["3-5 years"],["5"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["0-2 years"],["2"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["6-8 years"],["8"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["9-11 years"],["11"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["12-14 years"],["14"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["18-20 years"],["20"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["15-17 years"],["17"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["21-23 years"],["23"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["30 or more years"],["30"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["24-26 years"],["26"])
combined['YearsCodingProf']=combined['YearsCodingProf'].replace(["27-29 years"],["29"])

combined
```

```python
combined.to_csv('../Outputs/2018.csv', index=False)
```
