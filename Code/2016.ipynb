```python
import pandas as pd
```

```python
#import 2016 csv
data = pd.read_csv("../Resources/2016.csv")
data.head()
```

```python
data.columns
```

```python
# make df with only the important columns
cleaned = data[['experience_range','occupation','salary_range','company_size_range']]
cleaned['Year'] = '2016'
cleaned = cleaned.dropna()
```

```python
cleaned['occupation'].value_counts()
```

```python
# create smaller df with only the jobs we want
analyst = cleaned[cleaned['occupation'].str.contains('analyst')]
scientist = cleaned[cleaned['occupation'].str.contains('scientist')]
machine = cleaned[cleaned['occupation'].str.contains('machine')]
# engineer = cleaned[cleaned['DevType'].str.contains('engineer')]
```

```python
# convert job titles to specific
analyst['occupation'] = 'Data Analyst'
scientist['occupation'] = 'Data Scientist'
machine['occupation'] = 'Machine Learning Engineer'
```

```python
combined = pd.concat([analyst, scientist, machine], ignore_index=True)
combined
```

```python
combined['company_size_range'].value_counts()
```

```python
combined['company_size_range']=combined['company_size_range'].replace(["10,000+ employees"],["10000+"])
combined['company_size_range']=combined['company_size_range'].replace(["20-99 employees"],["10-100"])
combined['company_size_range']=combined['company_size_range'].replace(["100-499 employees"],["100-1000"])
combined['company_size_range']=combined['company_size_range'].replace(["1,000-4,999 employees"],["1000-4999"])
combined['company_size_range']=combined['company_size_range'].replace(["1-4 employees"],["2-9"])
combined['company_size_range']=combined['company_size_range'].replace(["10-19 employees"],["10-100"])
combined['company_size_range']=combined['company_size_range'].replace(["500-999 employees"],["100-1000"])
combined['company_size_range']=combined['company_size_range'].replace(["5,000-9,999 employees"],["5000-9999"])
combined['company_size_range']=combined['company_size_range'].replace(["5-9 employees"],["2-9"])
combined['company_size_range']=combined['company_size_range'].replace(["I am not sure"],["0"])
combined['company_size_range']=combined['company_size_range'].replace(["I am not part of a company"],["1"])



combined
```

```python
combined['experience_range'].value_counts()
```

```python
combined['experience_range']=combined['experience_range'].replace(["2 - 5 years"],["5"])
combined['experience_range']=combined['experience_range'].replace(["11+ years"],["15"])
combined['experience_range']=combined['experience_range'].replace(["6 - 10 years"],["10"])
combined['experience_range']=combined['experience_range'].replace(["1 - 2 years"],["2"])
combined['experience_range']=combined['experience_range'].replace(["Less than 1 year"],["1"])
combined
```
