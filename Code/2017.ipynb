```python
import pandas as pd
```

```python
#import 2017 csv
data = pd.read_csv("../Resources/2017.csv")
data.head()
```

```python
data.columns
```

```python
# make df with only the important columns
cleaned = data[['YearsCodedJob','DeveloperType','Salary','CompanySize']]
cleaned['Year'] = '2017'
cleaned = cleaned.dropna()
```

```python
cleaned['DeveloperType'].value_counts()
```

```python
# create smaller df with only the jobs we want
analyst = cleaned[cleaned['DeveloperType'].str.contains('analyst')]
scientist = cleaned[cleaned['DeveloperType'].str.contains('scientist')]
machine = cleaned[cleaned['DeveloperType'].str.contains('machine')]
# engineer = cleaned[cleaned['DevType'].str.contains('engineer')]
```

```python
# convert job titles to specific
analyst['DeveloperType'] = 'Data Analyst'
scientist['DeveloperType'] = 'Data Scientist'
machine['DeveloperType'] = 'Machine Learning Engineer'
```

```python
combined = pd.concat([analyst, scientist, machine], ignore_index=True)
combined
```

```python
combined['CompanySize'].value_counts()
```

```python
combined['CompanySize']=combined['CompanySize'].replace(["10,000 or more employees"],["10000+"])
combined['CompanySize']=combined['CompanySize'].replace(["20 to 99 employees"],["10-100"])
combined['CompanySize']=combined['CompanySize'].replace(["100 to 499 employees"],["100-1000"])
combined['CompanySize']=combined['CompanySize'].replace(["1,000 to 4,999 employees"],["1000-4999"])
combined['CompanySize']=combined['CompanySize'].replace(["2 to 9 employees"],["2-9"])
combined['CompanySize']=combined['CompanySize'].replace(["10 to 19 employees"],["10-100"])
combined['CompanySize']=combined['CompanySize'].replace(["500 to 999 employees"],["100-1000"])
combined['CompanySize']=combined['CompanySize'].replace(["5,000 to 9,999 employees"],["5000-9999"])
combined['CompanySize']=combined['CompanySize'].replace(["Fewer than 10 employees"],["2-9"])
combined['CompanySize']=combined['CompanySize'].replace(["I don’t know"],["0"])
combined['CompanySize']=combined['CompanySize'].replace(["I prefer not to answer"],["0"])



combined
```

```python
combined['YearsCodedJob'].value_counts()
```

```python
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["3 to 4 years"],["4"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["1 to 2 years"],["2"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["2 to 3 years"],["3"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["20 or more years"],["20"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["4 to 5 years"],["5"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["5 to 6 years"],["6"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["9 to 10 years"],["10"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["Less than a year"],["1"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["10 to 11 years"],["11"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["7 to 8 years"],["8"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["6 to 7 years"],["7"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["16 to 17 years"],["17"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["8 to 9 years"],["9"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["15 to 16 years"],["16"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["13 to 14 years"],["14"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["12 to 13 years"],["13"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["14 to 15 years"],["15"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["11 to 12 years"],["12"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["17 to 18 years"],["18"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["19 to 20 years"],["20"])
combined['YearsCodedJob']=combined['YearsCodedJob'].replace(["18 to 19 years"],["19"])

combined
```
