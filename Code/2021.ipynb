```python
import pandas as pd
```

```python
#import 2021 csv
data = pd.read_csv("../Resources/2021.csv")
data.head()
```

```python
data.columns
```

```python
# make df with only the important columns
cleaned = data[['YearsCodePro','DevType','ConvertedCompYearly','OrgSize']]
cleaned['Year'] = '2021'
cleaned = cleaned.dropna()
```

```python
cleaned['DevType'].value_counts()
```

```python
# create smaller df with only the jobs we want
analyst = cleaned[cleaned['DevType'].str.contains('analyst')]
scientist = cleaned[cleaned['DevType'].str.contains('scientist')]
machine = cleaned[cleaned['DevType'].str.contains('machine')]
# engineer = cleaned[cleaned['DevType'].str.contains('engineer')]
```

```python
# convert job titles to specific
analyst['DevType'] = 'Data Analyst'
scientist['DevType'] = 'Data Scientist'
machine['DevType'] = 'Machine Learning Engineer'
```

```python
combined = pd.concat([analyst, scientist, machine], ignore_index=True)
combined
```

```python
combined['OrgSize'].value_counts()
```

```python
combined['OrgSize']=combined['OrgSize'].replace(["10,000 or more employees"],["10000+"])
combined['OrgSize']=combined['OrgSize'].replace(["20 to 99 employees"],["10-100"])
combined['OrgSize']=combined['OrgSize'].replace(["100 to 499 employees"],["100-1000"])
combined['OrgSize']=combined['OrgSize'].replace(["1,000 to 4,999 employees"],["1000-4999"])
combined['OrgSize']=combined['OrgSize'].replace(["2 to 9 employees"],["2-9"])
combined['OrgSize']=combined['OrgSize'].replace(["10 to 19 employees"],["10-100"])
combined['OrgSize']=combined['OrgSize'].replace(["500 to 999 employees"],["100-1000"])
combined['OrgSize']=combined['OrgSize'].replace(["5,000 to 9,999 employees"],["5000-9999"])
combined['OrgSize']=combined['OrgSize'].replace(["Just me - I am a freelancer, sole proprietor, etc."],["1"])
combined['OrgSize']=combined['OrgSize'].replace(["I don’t know"],["0"])

combined
```

```python
combined.to_csv('../Outputs/2021.csv', index=False)
```
